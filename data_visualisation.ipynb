```python
import json
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Load JSON data
# -----------------------------
json_file = r"C:\Users\siauw\OneDrive\Desktop\Modelling and Visualisation\glauber_50_heating\Observables.json"
json_file_1 = r"C:\Users\siauw\OneDrive\Desktop\Modelling and Visualisation\kawasaki_50_cooling\Observables.json"

with open(json_file, "r") as f:
    data = json.load(f)

with open(json_file_1, "r") as f:
    data1 = json.load(f)

T_vals_glauber = np.array(data["T_vals"])
T_vals_kawasaki = np.array(data1["T_vals"])

# Energy and Heat Capacity
E = np.array(data["energy"]["values"])
E_err = np.array(data["energy"]["errors"])
Cv = np.array(data["heat_capacity"]["values"])
Cv_err = np.array(data["heat_capacity"]["errors"])

# Energy and Heat Capacity
E1 = np.array(data1["energy"]["values"])
E_err1 = np.array(data1["energy"]["errors"])
Cv1 = np.array(data1["heat_capacity"]["values"])
Cv_err1 = np.array(data1["heat_capacity"]["errors"])

# Magnetization and Susceptibility
M = np.array(data["magnetization"]["values"])
M_err = np.array(data["magnetization"]["errors"])
chi = np.array(data["susceptibility"]["values"])
chi_err = np.array(data["susceptibility"]["errors"])
```

```python
# -----------------------------
# Figure 1: Energy and Heat Capacity
# -----------------------------
fig1, axes1 = plt.subplots(1, 2, figsize=(15, 8), sharex=True)

# Subplot 1: Average Energy
axes1[0].errorbar(T_vals_glauber, E, yerr=E_err, fmt='o-', markersize=4, elinewidth=1.5, capsize=3, color='tab:blue', label='Glauber')
axes1[0].errorbar(T_vals_kawasaki, E1, yerr=E_err1, fmt='o-', markersize=4, elinewidth=1.5, capsize=3, color='tab:orange', label='Kawasaki')
axes1[0].set_ylabel(r"Average Energy  $\langle E \rangle$")
axes1[0].set_xlabel(r"Temperature [$k_bT$]")
axes1[0].set_title(r"Average Energy $\langle E \rangle$ against Temperature [$k_bT$]")
axes1[0].legend()

# Subplot 2: Heat Capacity
axes1[1].errorbar(T_vals_glauber, Cv, yerr=Cv_err, fmt='s-', markersize=4, elinewidth=1.5, capsize=3, color='tab:blue', label='Glauber')
axes1[1].errorbar(T_vals_kawasaki, Cv1, yerr=Cv_err1, fmt='s-', markersize=4, elinewidth=1.5, capsize=3, color='tab:orange', label='Kawasaki')
axes1[1].set_xlabel(r"Temperature [$k_bT$]")
axes1[1].set_ylabel(r"Heat Capacity $C_v$")
axes1[1].set_title(r"Heat Capacity [$C_v$] against Temperature [$k_bT$]")
axes1[1].legend()

plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Figure 2: Magnetization and Susceptibility
# -----------------------------
fig2, axes2 = plt.subplots(1, 2, figsize=(15, 8), sharex=True)

# Subplot 1: Average Magnetization
axes2[0].errorbar(T_vals_glauber, M, yerr=M_err, fmt='o-', markersize=4, elinewidth=1.5, capsize=3, color='tab:blue')
axes2[0].set_ylabel(r"Average Magnetization $\langle |M| \rangle$")
axes2[0].set_xlabel(r"Temperature [$k_bT$]")
axes2[0].set_title(r"Average Magnetization $\langle |M| \rangle$ against Temperature [$k_bT$]")

# Subplot 2: Susceptibility
axes2[1].errorbar(T_vals_glauber, chi, yerr=chi_err, fmt='s-', markersize=4, elinewidth=1.5, capsize=3, color='tab:blue')
axes2[1].set_xlabel(r"Temperature [$k_bT$]")
axes2[1].set_ylabel(r"Susceptibility $\chi$")
axes2[1].set_title(r"Magnetic Susceptibility [$\chi$] against Temperature [$k_bT$]")

plt.tight_layout()
plt.show()
```
